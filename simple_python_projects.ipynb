Assignment 1: Temperature Check

```python
temperature = float(input("enter temperature: "))

if temperature < 10:
    print("It's too cold!")
elif temperature <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

Assignment 2: ATM Withdrawal System

```python
balance = float(input("enter account balance: "))
withdraw = float(input("enter withdrawal amount: "))

if withdraw>balance:
    print("Insufficient Balance")
else:
    print("Transaction Successful")
    print("Remaining Balance:",balance-withdraw)
```

Assignment 3: Student Grade Calculator

```python
marks = int(input("enter marks(0–100): "))

if marks >= 90 and marks <= 100:
    print("A+")
elif marks >= 80:
    print("A")
elif marks >= 70:
    print("B")
elif marks >= 60:
    print("C")
else:
    print("Fail")
```

Assignment 4: Bus Ticket Discount

```python
age = int(input("enter age: "))

if age < 5:
    print("Free")
elif age <= 18:
    print("50% Discount")
elif age >= 60:
    print("30% Discount")
else:
    print("Full Price")
```

Assignment 5: Restaurant Bill with Discount

```python
bill = float(input("enter total bill: "))

if bill > 1000:
    discount = bill*0.10
    bill = bill-discount
    print("10% discount")
else:
    print("no discount")

print("final bill:",bill)
```

Assignment 7: Traffic Light System

```python
color = input("enter traffic light color (red, yellow, green): ")

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color")
```

Assignment 8: Leap Year Checker

```python
year = int(input("enter a year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

Assignment 9: Job Eligibility Checker

Assignment 10: Electricity Bill Calculator

```python
units = int(input("enter number of units consumed: "))

if units <= 100:
    bill = units*5
elif units <= 200:
    bill = units*7
else:
    bill = units*10

print("total amount:",bill,"BDT")
```

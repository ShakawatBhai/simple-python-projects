1. Inventory Price Calculator
Real-world use: Used in e-commerce dashboards for price summaries.
Task:
Write a function calculate_inventory_summary (prices) that:

 receives a list of product prices

 returns total_cost and average_cost

Example:

Input: [120, 250, 399, 150]

Output: (919, 229.75)

```python
def calculate_inventory_summary(prices):
    prices = [float(p) for p in prices]
    if not prices:
        return 0, 0.0
    total = sum(prices)
    avg = total/len(prices)
    return total,avg


Innput = input("Enter prices (like [120, 250, 399, 150]): ")
main_list = Innput.strip('[]').split(',')
main = [item.strip() for item in main_list]

total,avg = calculate_inventory_summary(main)

print(f"({total},{avg})")
```

2. Unique Visitor Tracker
Real-world use: Website analytics tools calculating daily active users.

Task:
Write a function count_unique_visitors(visitor_list) that:

 takes a list of visitor IDs

 returns the number of unique visitors

Example:

Input: [101, 205, 101, 310, 205, 550]

Output: 4 (The unique IDs are 101, 205, 310, 550)

```python
def count_unique_visitors(visitor_list):
    # Convert the list to a set to get only unique visitor IDs
    unique_visitors = set(visitor_list)
    # Return the number of unique visitors (the length of the set) and the set itself
    return len(unique_visitors), unique_visitors

# Example usage:
# Take input from the user for visitor IDs
input_str = input("Enter visitor IDs (like [101, 205, 101, 310, 205, 550]): ")

# Process the input string
# Remove brackets and split by comma
raw_list_str = input_str.strip('[]').split(',')
# Convert each string ID to an integer
visitor_ids = [int(item.strip()) for item in raw_list_str if item.strip()]

# Call the function with the processed list
num_unique, unique_ids = count_unique_visitors(visitor_ids)

# Print the result, including the unique IDs
print(f"The number of unique visitors is: {num_unique} (The unique IDs are {sorted(list(unique_ids))})")
```

3. Product Category Counter
Real-world use: Inventory/category-based analytics.
Task:

Write a function count_categories(products) that:

 takes a list of product names

 returns a dictionary counting each product type.

Example:
Input: ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]

Output: {"Laptop": 3, "Smartphone": 2, "Monitor": 1, "Headphones": 1}

```python
from collections import Counter

def count_categories(products):
    # Use collections.Counter to count occurrences of each product
    return Counter(products)

# Example usage:
# Take input from the user for product names
input_str = input("Enter product names (like [\"Laptop\", \"Smartphone\", \"Monitor\"]): ")

# Process the input string
# Remove brackets and split by comma, then strip whitespace and quotes
raw_list_str = input_str.strip('[]').split(',')
products_list = [item.strip().strip('"') for item in raw_list_str if item.strip()]

# Call the function with the processed list
product_counts = count_categories(products_list)

# Print the result
print(product_counts)
```

4. Temperature Analytics
Real-world use: Health tech temperature screening systems.
Task:

Write a function analyze_temperatures(temp_list) that:

 separates high (> 38°C) and normal temps

 returns two lists: (high_list, normal_list)

Example:

Input: [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

Output:
High List (>38°C): [39.1, 40.5]
Normal List (<38°C): [36.5, 38.0, 37.2, 38.0, 37.9]

Full Output Tuple: ([39.1, 40.5], [36.5, 38.0, 37.2, 38.0, 37.9])

```python
def analyze_temperatures(temp_list):
    high_list = []
    normal_list = []
    for temp in temp_list:
        if temp > 38.0:
            high_list.append(temp)
        else:
            normal_list.append(temp)
    return high_list, normal_list

# Example usage:
# Take input from the user for temperatures
input_str = input("Enter temperatures (like [36.5, 38.0, 39.1]): ")

# Process the input string
# Remove brackets and split by comma
raw_list_str = input_str.strip('[]').split(',')
# Convert each string to a float
temperatures = [float(item.strip()) for item in raw_list_str if item.strip()]

# Call the function with the processed list
high_temps, normal_temps = analyze_temperatures(temperatures)

# Print the results in the specified format
print(f"High List (>38°C): {high_temps}")
print(f"Normal List (<=38°C): {normal_temps}")
print(f"Full Output Tuple: ({high_temps}, {normal_temps})")
```

5. Menu Price Lookup
Real-world use: Restaurant ordering apps.

Task:
Write a function get_price(menu, item) that:

 receives a dictionary menu and an item name

 returns the price if exists

 otherwise returns "Item not found"

Example 1: Item Found

 Input (Menu): {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}

 Input (Item): "Muffin"

 Output: 4.00
Example 2: Item Not Found

 Input (Menu): {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}

 Input (Item): "Donut"

 Output: "Item not found"

```python
import ast

def get_price(menu, item):
    if item in menu:
        return menu[item]
    else:
        return "Item not found"

# Take input from the user for the menu
menu_input_str = input("Enter menu as a dictionary (e.g., {'Coffee': 3.50, 'Tea': 2.75}): ")

try:
    # Safely evaluate the string input as a dictionary
    user_menu = ast.literal_eval(menu_input_str)
    if not isinstance(user_menu, dict):
        raise ValueError("Input is not a valid dictionary.")
except (ValueError, SyntaxError) as e:
    print(f"Invalid menu input: {e}")
    user_menu = {}

# Take input from the user for the item
item_input = input("Enter the item name to look up (e.g., 'Muffin'): ")
# Strip leading/trailing whitespace and then any enclosing quotes (both single and double)
item_input = item_input.strip().strip('"').strip('\'')

# Call the function with user inputs
output = get_price(user_menu, item_input)

print(f"Input (Menu): {user_menu}\nInput (Item): \"{item_input}\"\nOutput: {output}")
```

6. Student Grading System
Real-world use: School result automation.
Task:

Write a function calculate_grade(score) that:

 returns "A+", "A", "A-", "F" based on numeric marks
Write another function grade_students(student_dict) to:

 take a dictionary of student:score

 return student:grade dictionary

Example:

Input (Scores Dictionary): {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}

Output (Grades Dictionary): {"Alice": "A+", "Bob": "A", "Charlie": "F", "David": "A-"}

```python
import ast

def calculate_grade(score):
    if score >= 85:
        return "A+"
    elif score >= 75:
        return "A"
    elif score >= 65:
        return "A-"
    else:
        return "F"

def grade_students(student_dict):
    graded_students = {}
    for student, score in student_dict.items():
        graded_students[student] = calculate_grade(score)
    return graded_students

# Take input from the user for the scores dictionary
scores_input_str = input("Enter student scores as a dictionary (e.g., {'Alice': 85, 'Bob': 78}): ")

try:
    # Safely evaluate the string input as a dictionary
    user_scores_dict = ast.literal_eval(scores_input_str)
    if not isinstance(user_scores_dict, dict):
        raise ValueError("Input is not a valid dictionary.")
except (ValueError, SyntaxError) as e:
    print(f"Invalid scores input: {e}")
    user_scores_dict = {}

# Call the function with the processed dictionary
grades_output = grade_students(user_scores_dict)

print(f"Output (Grades Dictionary): {grades_output}")
```

7. Discount Calculator
Real-world use: Online store discount engine.

Task:
Write a function apply_discounts(products) where:

 products is a list of tuples: (name, price, discount)

 return a list of updated prices after discount
Example:

Input (List of Tuples): [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]

Output (List of Prices): [64.00, 22.50, 285.00]

```python
import ast

def apply_discounts(products):
    updated_prices = []
    for name, price, discount in products:
        discount_amount = price * (discount / 100)
        final_price = price - discount_amount
        updated_prices.append(round(final_price, 2)) # Round to 2 decimal places for currency
    return updated_prices

# Example usage:
# Take input from the user for the list of products
input_str = input('Enter products as a list of tuples (e.g., [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10)]): ')

try:
    # Safely evaluate the string input as a list of tuples
    user_products = ast.literal_eval(input_str)
    # Basic validation to ensure it's a list of tuples with 3 elements
    if not isinstance(user_products, list) or \
       not all(isinstance(item, tuple) and len(item) == 3 for item in user_products):
        raise ValueError("Input is not a valid list of (name, price, discount) tuples.")
except (ValueError, SyntaxError) as e:
    print(f"Invalid input: {e}")
    user_products = [] # Fallback to an empty list on error

# Call the function with the processed list
if user_products:
    output_prices = apply_discounts(user_products)
    print(f"Output (List of Prices): {output_prices}")
else:
    print("No valid products entered.")
```

8. Hashtag Generator
Real-world use: Social media automation tools.

Task:
Write a function generate_hashtags(sentence) that:

 converts every word into a hashtag

 returns a list of hashtags

Example:
Input: "AI Data Science"

Output: ["#ai", "#data", "#science"]

```python
def generate_hashtags(sentence):
    # Split the sentence into words, convert to lowercase, and prefix with '#'
    words = sentence.split()
    hashtags = [f"#{word.lower()}" for word in words]
    return hashtags

# Example usage:
# Take input from the user for the sentence
input_sentence = input("Enter a sentence (e.g., 'AI Data Science'): ")

# Strip any enclosing quotes from the input sentence
input_sentence = input_sentence.strip('"').strip("'")

# Call the function with the user's input
output_hashtags = generate_hashtags(input_sentence)

# Print the result
print(f"Input: \"{input_sentence}\"\nOutput: {output_hashtags}")
```

9. User Authentication Mock
Real-world use: Login systems (basic level).
Task:

Write a function login(username, password) that:

 checks if username exists
 checks if password matches
 returns "Login Successful" or "Invalid Credentials"

Example Setup (Mock Database): Assume the following credential database is used internally:
CREDENTIALS = {"alex_s": "securepwd1", "maria_t": "password101", "john_d": "johndoe123"}


Example 1: Successful Login
 Input (Username): "maria_t"

 Input (Password): "password101"

 Output: "Login Successful"
Example 2: Failed Login (Wrong Password)

 Input (Username): "alex_s"

 Input (Password): " securepwd12"

 Output: "Invalid Credentials"

Example 3: Failed Login (Non-existent User)

 Input (Username): "peter_z"

 Input (Password): "any_password"

 Output: "Invalid Credentials"

```python
CREDENTIALS = {"alex_s": "securepwd1", "maria_t": "password101", "john_d": "johndoe123"}

def login(username, password):
    if username in CREDENTIALS:
        if CREDENTIALS[username] == password:
            return "Login Successful"
        else:
            return "Invalid Credentials" # Correct username, but wrong password
    else:
        return "Invalid Credentials" # Username does not exist

# Take input from the user for username
user_input_username = input("Enter username: ")
# Strip leading/trailing whitespace and then any enclosing quotes (both single and double)
user_input_username = user_input_username.strip().strip('"').strip('\'')

# Take input from the user for password
user_input_password = input("Enter password: ")
# Strip leading/trailing whitespace and then any enclosing quotes (both single and double)
user_input_password = user_input_password.strip().strip('"').strip('\'')

# Call the login function with user inputs
login_status = login(user_input_username, user_input_password)

print(f"Input (Username): \"{user_input_username}\"\nInput (Password): \"{user_input_password}\"\nOutput: {login_status}")
```

10. Shopping Cart Bill Calculator

Real-world use: Billing systems in e-commerce apps.

Task:

Write a function calculate_total(cart, prices) that:

 takes a shopping cart list

 takes a dictionary of item:price

 returns the total bill amount

Example:

Input (Cart): ["Milk", "Bread", "Milk", "Cheese", "Bread"]

Input (Price): {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}

Output (Total Bill Amount): 16.00

```python
import ast
from collections import Counter

def calculate_total(cart, prices):
    total_bill = 0.0
    for item in cart:
        if item in prices:
            total_bill += prices[item]
        else:
            print(f"Warning: Item '{item}' not found in price list. Skipping this item.")
    return total_bill

# Example usage:
# Take input from the user for the shopping cart
cart_input_str = input('Enter shopping cart items as a list (e.g., ["Milk", "Bread", "Milk"]): ')

try:
    # Safely evaluate the string input as a list
    user_cart = ast.literal_eval(cart_input_str)
    if not isinstance(user_cart, list):
        raise ValueError("Input is not a valid list.")
except (ValueError, SyntaxError) as e:
    print(f"Invalid cart input: {e}")
    user_cart = []

# Take input from the user for the prices dictionary
prices_input_str = input("Enter prices as a dictionary (e.g., {'Milk': 3.00, 'Bread': 2.50}): ")

try:
    # Safely evaluate the string input as a dictionary
    user_prices = ast.literal_eval(prices_input_str)
    if not isinstance(user_prices, dict):
        raise ValueError("Input is not a valid dictionary.")
except (ValueError, SyntaxError) as e:
    print(f"Invalid prices input: {e}")
    user_prices = {}

# Call the function with user inputs
if user_cart and user_prices:
    total_bill_amount = calculate_total(user_cart, user_prices)
    print(f"Input (Cart): {user_cart}")
    print(f"Input (Price): {user_prices}")
    print(f"Output (Total Bill Amount): {total_bill_amount:.2f}")
else:
    print("Cannot calculate total bill due to invalid input.")
```

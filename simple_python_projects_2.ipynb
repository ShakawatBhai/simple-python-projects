Email Validator (String Filtering + Loops)
ðŸ’¼ Real-world use: Used in signup forms or data cleaning tasks.

Task:
Check if a given email is valid â€” must contain â€œ@â€ and end with â€œ.comâ€ or â€œ.orgâ€.


Example:
Input: hello@inceptionbd.com

Output: Valid Email

```python
email = input("Enter email: ")

valid = True
at_count = 0
at_index = -1

for i in range(len(email)):
    if email[i] == '@':
        at_count += 1
        at_index = i

if at_count != 1 or at_index == 0 or at_index == len(email) - 1:
    valid = False

end1 = ".com"
end2 = ".org"

match_com = True
match_org = True

if len(email) >= len(end1):
    for i in range(1, len(end1) + 1):
        if email[-i] != end1[-i]:
            match_com = False
            break
else:
    match_com = False

if len(email) >= len(end2):
    for i in range(1, len(end2) + 1):
        if email[-i] != end2[-i]:
            match_org = False
            break
else:
    match_org = False

if not (match_com or match_org):
    valid = False

if valid:
    print("Valid Email")
else:
    print("Invalid Email")
```

Count Word Frequency in Customer Reviews
ðŸ’¼ Real-world use: Analyzing reviews or comments in text analytics.

Task:
Given a review text, count how many times each word appears (ignore case).

Example:
Input: "This product is good and this service is excellent"

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
input_text = input("Enter your review text: ")

cleaned_text = ""
for char in input_text:
    if 'a' <= char.lower() <= 'z' or char == ' ':
        cleaned_text += char.lower()

words = cleaned_text.split()

word_counts = {}
for word in words:
    if word:
        word_counts[word] = word_counts.get(word, 0) + 1

print(word_counts)
```

3. Password Strength Checker
ðŸ’¼ Real-world use: Used in user authentication systems.

Task:
Check if a password is strong â€” must have at least one uppercase, one lowercase, one digit, and one
special character.

Example:
Input: "Hello@123"

Output: Strong Password

```python
password = input("Enter your password: ")

has_uppercase = False
has_lowercase = False
has_digit = False
has_special = False

for char in password:
    if 'A' <= char <= 'Z':
        has_uppercase = True
    elif 'a' <= char <= 'z':
        has_lowercase = True
    elif '0' <= char <= '9':
        has_digit = True
    else:
        if not ('A' <= char <= 'Z' or 'a' <= char <= 'z' or '0' <= char <= '9'):
            has_special = True

if has_uppercase and has_lowercase and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

SMS Word Counter
ðŸ’¼ Real-world use: Used in messaging apps or social media post limit check.

Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

Example:
Input: "This is a demo message."

Output: Words = 5, Characters = 23, Status = Within Limit

```python
message = input("Enter your message: ")

character_count = 0
for char in message:
    character_count += 1

word_count = 0
is_word = False
for char_index in range(character_count):
    if message[char_index] != ' ':
        if not is_word:
            word_count += 1
            is_word = True
    else:
        is_word = False

status = ""
if character_count <= 160:
    status = "Within Limit"
else:
    status = "Exceeds Limit"

print("Words = ", word_count, ", Characters = ", character_count, ", Status = ", status)
```

Remove Duplicates from Contact List
ðŸ’¼ Real-world use: Data cleaning in CRM or user databases.

Task:
Given a list of names (comma separated), remove duplicates.

Example:
Input: "Rahim, Karim, Rahim, Sultana"

Output: "Rahim, Karim, Sultana"

```python
names_input = input("Enter names (comma separated): ")

cleaned_input_string = ""
if len(names_input) > 0 and names_input[0] == '"' and names_input[-1] == '"':
    for i in range(1, len(names_input) - 1):
        cleaned_input_string += names_input[i]
else:
    cleaned_input_string = names_input

unique_names_ordered = []
seen_names = []

current_name_buffer = ""

for i in range(len(cleaned_input_string)):
    char = cleaned_input_string[i]

    if char == ',':
        trimmed_name = ""
        start_index = 0
        end_index = len(current_name_buffer) - 1

        while start_index < len(current_name_buffer) and current_name_buffer[start_index] == ' ':
            start_index += 1
        while end_index >= start_index and current_name_buffer[end_index] == ' ':
            end_index -= 1

        for j in range(start_index, end_index + 1):
            trimmed_name += current_name_buffer[j]

        if trimmed_name != "":
            is_seen = False
            for seen in seen_names:
                if seen == trimmed_name:
                    is_seen = True
                    break
            if not is_seen:
                unique_names_ordered.append(trimmed_name)
                seen_names.append(trimmed_name)

        current_name_buffer = ""
    else:
        current_name_buffer += char

if current_name_buffer != "":
    trimmed_name = ""
    start_index = 0
    end_index = len(current_name_buffer) - 1

    while start_index < len(current_name_buffer) and current_name_buffer[start_index] == ' ':
        start_index += 1
    while end_index >= start_index and current_name_buffer[end_index] == ' ':
        end_index -= 1

    for j in range(start_index, end_index + 1):
        trimmed_name += current_name_buffer[j]

    if trimmed_name != "":
        is_seen = False
        for seen in seen_names:
            if seen == trimmed_name:
                is_seen = True
                break
        if not is_seen:
            unique_names_ordered.append(trimmed_name)
            seen_names.append(trimmed_name)

output_string = ""
for i in range(len(unique_names_ordered)):
    output_string += unique_names_ordered[i]
    if i < len(unique_names_ordered) - 1:
        output_string += ", "

print(output_string)
```

Real-world use: Used in chat moderation or email filtering.

Task:
If any â€œbannedâ€ words like ['spam', 'scam', 'fake'] are found in a text, print â€œWarningâ€.

Example:
Input: "This is a scam offer"

Output: Warning: Suspicious content detected

```python
message_input = input("Enter your message: ")

banned_words = ['spam', 'scam', 'fake']

cleaned_message = ""
for char in message_input:
    if 'a' <= char.lower() <= 'z' or char == ' ':
        cleaned_message += char.lower()

words_in_message = []
current_word = ""
for char in cleaned_message:
    if char == ' ':
        if current_word:
            words_in_message.append(current_word)
        current_word = ""
    else:
        current_word += char
if current_word:
    words_in_message.append(current_word)

is_suspicious = False
for msg_word in words_in_message:
    for banned_word in banned_words:
        if msg_word == banned_word:
            is_suspicious = True
            break
    if is_suspicious:
        break

if is_suspicious:
    print("Warning: Suspicious content detected")
else:
    print("No suspicious content found.")
```

Invoice Code Formatter
ðŸ’¼ Real-world use: Formatting IDs or codes in finance or ERP systems.

Task:
Given an invoice number like 23, generate code in the format INV00023.

Example:
Input: 23

Output: INV00023

```python
invoice_number_str = input("Enter invoice number: ")

invoice_number = 0
for char in invoice_number_str:
    digit = ord(char) - ord('0')
    invoice_number = invoice_number * 10 + digit

target_digit_length = 5

current_num_str = ""
if invoice_number == 0:
    current_num_str = "0"
else:
    temp_num = invoice_number
    while temp_num > 0:
        digit = temp_num % 10
        current_num_str = chr(ord('0') + digit) + current_num_str
        temp_num = temp_num // 10

zeros_to_add = target_digit_length - len(current_num_str)

padded_number_str = ""
for _ in range(zeros_to_add):
    padded_number_str += "0"
padded_number_str += current_num_str

formatted_invoice_code = "INV" + padded_number_str

print(formatted_invoice_code)
```

Word Reversal for Data Encryption
ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding.

Task:
Reverse each word in a sentence while keeping word order the same.

Example:
Input: "AI is fun"

Output: "IA si nuf"

```python
sentence_input_raw = input("Enter your sentence: ")

sentence_input = ""
if len(sentence_input_raw) > 0 and sentence_input_raw[0] == '"' and sentence_input_raw[-1] == '"':
    for i in range(1, len(sentence_input_raw) - 1):
        sentence_input += sentence_input_raw[i]
else:
    sentence_input = sentence_input_raw

words = []
current_word = ""

for char in sentence_input:
    if char == ' ':
        if current_word:
            words.append(current_word)
        current_word = ""
    else:
        current_word += char

if current_word:
    words.append(current_word)

reversed_words = []

for word in words:
    reversed_word = ""
    for i in range(len(word) - 1, -1, -1):
        reversed_word += word[i]
    reversed_words.append(reversed_word)

output_sentence = ""
for i in range(len(reversed_words)):
    output_sentence += reversed_words[i]
    if i < len(reversed_words) - 1:
        output_sentence += " "
print(f'"{output_sentence}"')
```
